```python
class bill():
    menu = {'Dosa':50,'idli':30,'chawal dal':60}

    def order(self):
        order = {}
        n = int(input("Enter no. of items: "))
        for i in range(n):
            item = input("Enter item: ")
            if item in self.menu:
                qty = int(input(f"Enter quantity of {item}: "))
                order[item] = [qty, self.menu[item]]
            else:
                print("Item not available")
        return order

    def printBill(self, order):
        grand_total = 0
        sno = 1

        # Open file in write mode
        with open("/content/drive/MyDrive/BillPrinting.txt", "w") as file:

            file.write("-"*50 + "\n")
            file.write("{:|>1}{:^50}{:|<1}\n".format("", "BILL", ""))
            file.write("-"*50 + "\n")
            file.write("{:|>1}{:<10}{:<10}{:<10}{:<10}{:<10}{:|<1}\n"
                       .format("", 'Sno.', "Item", "Quantity", 'Price', 'Total', ""))

            for item, value in order.items():
                qty, price = value
                total = qty * price
                grand_total += total

                file.write("{:|>1}{:<10}{:<10}{:<10}{:<10}{:<10}{:|<1}\n"
                           .format("", sno, item, qty, price, total, ""))

                sno += 1

            file.write("-"*50 + "\n")
            file.write("{:|>1}{:<40}{:<10}{:|>1}\n"
                       .format("", "Grand Total", grand_total, ""))
            file.write("-"*50 + "\n")

        print(" Bill saved successfully in receipt.txt")


# Run program
obj = bill()
my_order = obj.order()
obj.printBill(my_order)
```
