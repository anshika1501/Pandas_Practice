```python
class bill():
  menu={'Dosa':50,'idli':30,'chawal dal':60}
  def order(self):
    order={}
    n=int(input("enter no. of items"))
    for i in range(n):
      item=input("Enter item")
      if item in self.menu:
        qty=int(input(f"Enter quantity of {item}:"))
        order[item]=[qty,self.menu[item]]
      else:
        print("item not available")
    return order
  def printBill(self,order):
    print("-"*50)
    print("{:|>1}{:^50}{:|<1}".format("","BILL",""))
    print("-"*50)
    print("{:|>1}{:<10}{:<10}{:<10}{:<10}{:<10}{:|<1}".format("",'Sno.',"Item","Quantity",'Price','Total',""))

    sno = 1
    grand_total = 0

    for item, value in order.items():
        qty, price = value
        total = qty * price
        grand_total += total

        print("{:|>1}{:<10}{:<10}{:<10}{:<10}{:<10}{:|<1}".format("",
            sno, item, qty, price, total,""))

        sno += 1
    print("-"*50)
    print("{:|>1}{:<40}{:<10}{:|>1}".format("","Grand Total", grand_total,""))
    print("-"*50)

obj=bill()
my_order=obj.order()
obj.printBill(my_order)
```
